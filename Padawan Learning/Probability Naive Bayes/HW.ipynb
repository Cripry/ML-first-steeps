```python
def normalize(x):
    """
    Function for remove nondesired characters from data
    """
    if type(x) == str:
        x = ''.join(i for i in x.split() if i.isalnum())
        if not x:
            x = np.nan
    return x
```

```python
# Initializing Models 
logreg = LogisticRegression(max_iter=1000)
linreg = LinearRegression()
```

```python
# Get values for missing data from each transformer
X_train_si_mean = si_mean.transform(X_train)
X_train_si_median = si_median.transform(X_train)
X_train_si_most_frequent = si_most_frequent.transform(X_train)
X_train_si_constant = si_constant.transform(X_train)
X_train_knni = knni.transform(X_train)

X_test_si_mean = si_mean.transform(X_test)
X_test_si_median = si_median.transform(X_test)
X_test_si_most_frequent = si_most_frequent.transform(X_test)
X_test_si_constant = si_constant.transform(X_test)
X_test_knni = knni.transform(X_test)

# Creating a list with all transformers for better accesing 
transformed_data_X_train = [X_train_si_mean,X_train_si_median,X_train_si_most_frequent,X_train_si_constant,X_train_knni]
transformed_data_X_test = [X_test_si_mean,X_test_si_median,X_test_si_most_frequent,X_test_si_constant,X_test_knni]

# Creating list with used technique in string format 
technique_list = ['si_mean', 'si_median', 'si_most_frequent', 'si_constant', 'knni']
```

```python
# Creating Table
TABLE = {'Model':[], 'Technique':[], 'Accuracy':[]}
```

```python
# Populating Table using Logistic Regression
for X_train, X_test, technique in zip(transformed_data_X_train, transformed_data_X_test, technique_list):
    TABLE['Model'].append('Logistic Regression')
    TABLE['Technique'].append('technique')
    logreg.fit(X_train, y_train)
    TABLE['Accuracy'].append(logreg.score(X_test, y_test))
```

```python
# Populating Table using Linear Regression
for X_train, X_test, technique in zip(transformed_data_X_train, transformed_data_X_test, technique_list):
    TABLE['Model'].append('Logistic Regression')
    TABLE['Technique'].append('technique')
    linreg.fit(X_train, y_train)
    TABLE['Accuracy'].append(linreg.score(X_test, y_test))
```
