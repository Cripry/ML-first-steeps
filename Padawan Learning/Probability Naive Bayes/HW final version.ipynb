```python
# Read dataset 
df = pd.read_csv('kidney_disease.csv')
```

```python
# Describe dataset
df
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
def normalize(x):
    """
    Function for remove nondesired characters from data
    """
    if type(x) == str:
        try:
            x = float(x)
        except Exception as e:
            x = ''.join(i for i in x.split() if i.isalnum())
        if not x:
            x = np.nan
    return x
```

```python
# Making maping of columns
maping_y = {'notckd': 0, 'ckd': 1}
```

```python
# Mapping and cleaning data
y = y.apply(lambda x: maping_y[normalize(x)])
for column in X.columns:
    X[column].apply(lambda x: normalize(x))
```

```python
# Normalize 
for column in X.columns:
    array_df = X[column]
    new_array = array_df.apply(normalize)
    X[column] = new_array
```

```python
categorical_columns = ['rbc', 'pc', 'ba', 'pcc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'wc']
```

```python
X = X.drop(categorical_columns, axis=1)
```

```python
X
```

```python
from sklearn.impute import SimpleImputer
```

```python
# Create an instance of the SimpleImputer with the strategy set as median.
from sklearn.impute import SimpleImputer
si_mean = SimpleImputer(strategy='mean')
si_median = SimpleImputer(strategy='median')
si_most_frequent = SimpleImputer(strategy='most_frequent')
si_constant = SimpleImputer(strategy='constant', fill_value=0)
```

```python
si_mean_X_data = si_mean.fit_transform(X)
si_median_X_data = si_median.fit_transform(X)
si_most_frequent_X_data = si_most_frequent.fit_transform(X)
si_constant_X_data = si_constant.fit_transform(X)
```

```python
# Importing Models
from sklearn.linear_model import LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
# Initializing Models
knn = KNN()
lr = LogisticRegression(max_iter=5000)
gnb = GaussianNB()
```

```python
# Creating Table
Table = {'Model':[], 'Technique':[], 'Accuracy':[]}
```

```python
# si_mean

# Split data in test and train part for si_mean
X_train, X_test, y_train, y_test = train_test_split(si_mean_X_data, y)


lr.fit(X_train,y_train)
knn.fit(X_train,y_train)
gnb.fit(X_train,y_train)

lr_score = lr.score(X_test, y_test)
knn_score = knn.score(X_test, y_test)
gnb_score = gnb.score(X_test, y_test)

Table['Model'].append('Linear Regression')
Table['Model'].append('KNN')
Table['Model'].append('Naive Bayes')

Table['Technique'].append('si_mean')
Table['Technique'].append('si_mean')
Table['Technique'].append('si_mean')

Table['Accuracy'].append(lr_score)
Table['Accuracy'].append(knn_score)
Table['Accuracy'].append(gnb_score)
```

```python
# si_median

# Split data in test and train part for si_mean
X_train, X_test, y_train, y_test = train_test_split(si_median_X_data, y)


lr.fit(X_train,y_train)
knn.fit(X_train,y_train)
gnb.fit(X_train,y_train)

lr_score = lr.score(X_test, y_test)
knn_score = knn.score(X_test, y_test)
gnb_score = gnb.score(X_test, y_test)

Table['Model'].append('Linear Regression')
Table['Model'].append('KNN')
Table['Model'].append('Naive Bayes')

Table['Technique'].append('si_median')
Table['Technique'].append('si_median')
Table['Technique'].append('si_median')

Table['Accuracy'].append(lr_score)
Table['Accuracy'].append(knn_score)
Table['Accuracy'].append(gnb_score)
```

```python
# si_most_frequent

# Split data in test and train part for si_mean
X_train, X_test, y_train, y_test = train_test_split(si_most_frequent_X_data, y)


lr.fit(X_train,y_train)
knn.fit(X_train,y_train)
gnb.fit(X_train,y_train)

lr_score = lr.score(X_test, y_test)
knn_score = knn.score(X_test, y_test)
gnb_score = gnb.score(X_test, y_test)

Table['Model'].append('Linear Regression')
Table['Model'].append('KNN')
Table['Model'].append('Naive Bayes')

Table['Technique'].append('si_most_frequent')
Table['Technique'].append('si_most_frequent')
Table['Technique'].append('si_most_frequent')

Table['Accuracy'].append(lr_score)
Table['Accuracy'].append(knn_score)
Table['Accuracy'].append(gnb_score)
```
