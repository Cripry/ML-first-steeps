```python
import pandas as pd
```

```python
#Read DataFrame
df = pd.read_csv('Materiale/beer_profile_and_ratings.csv')
```

```python
#Describe DataFrame
df.describe()
```

```python
#Get correlation between all columns
df.corr()
```

```python
# Separating Data from results
y = df['review_overall']
X = df.drop(['review_overall', 'Beer Name (Full)', 'Description','Name', 'Style','Brewery', 'review_aroma', 'review_palate', 'review_taste'], axis=1)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                   test_size=0.3,
                                                   random_state=42)
```

```python
#Create an instance of the LinearRegression class from sklearn.
lr = LinearRegression()
#Call the fit function, which applies the Linear Regression model
#onto the data
lr.fit(X_train, y_train)
#Define a variable that stores the predictions of our model
y_pred = lr.predict(X_test)

#Print the mean squares error and meand absolute error to get an idea
#how well the model has done.
mean_squared_error(y_pred, y_test), abs(mean_squared_error(y_pred, y_test))
```

```python
lr.score(X, y)
```
