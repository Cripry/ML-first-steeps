```python
# import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.datasets import mnist
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Importing data
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

# Select random interval of numbers
num = random.randint(0, len(X_train)-10)
images = X_train[num:num+10]
labels = Y_train[num:num+10]
```

```python
# 2 Second plot in a multiple window plot 10 random digits with a title having the digit from the image on a 2 x 5 grid.
num_row = 2
num_col = 5
# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for i in range(10):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(images[i], cmap='gray')
    ax.set_title(f'Label: {labels[i]}')
plt.tight_layout()
plt.show()
```

```python
# 3 Create an instance of the sklearn PCA.

pca = PCA(n_components=2)
scaler = StandardScaler()
```

```python
# Reshape Training value
X_train = X_train.reshape(60000, 28*28)
```

```python
# 4 Apply scaling on the digit data set.
scaler.fit(X_train)
scaled_data = scaler.transform(X_train)
```

```python
# 5 Train the PCA on the digit data set.
pca.fit(scaled_data)
pca_X = pca.transform(scaled_data)
```

```python
# 7 Plot the new result data set in a scatter plot with each point having colored by its digit label.
import matplotlib.pyplot as plt

fig = plt.gcf()

# Change seaborn plot size
fig.set_size_inches(19, 15)

# Ploting numbers and
import seaborn as sns 
sns.scatterplot(pca_X[:, 0], pca_X[:, 1], hue=Y_train, palette='tab10')
```
